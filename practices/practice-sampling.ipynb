**결측치 확인**

데이터셋의 크기인 891보다 적은 수의 `non-null` 값을 갖는 특성에 결측치가 .

**`Age` 특성 결측치 처리 방법 1: 특성 중앙값 활용**

`Age` 특성의 결측치를 다양한 방식으로 채운다.
따라서 `titanic` 데이터셋의  원본을 그대로 두고 복제해서 사용한다.

```python
titanic_median = titanic.copy()
```

`Age` 특성의 결측치를 모두 해당 특성의 중앙값으로 대체한다.

```python
age_median =titanic_median['Age'].median()
age_median
```

```python
titanic_median['Age'] = titanic_median['Age'].fillna(age_median)
```

```python
titanic_median.info()
```

**`Age` 특성 결측치 처리 방법 2: 성(Sex)별 중앙값 활용**

- 방식 1: 부울 인덱싱 활용

먼저 타이타닉 데이터셋을 복제한다.

```python
titanic_sex_median = titanic.copy()
```

여성과 남성의 중위연령을 확인한다.

```python
f_mask = titanic_sex_median["Sex"]=="female"
f_age_median = titanic_sex_median.loc[f_mask, "Age"].median()
print("여성 평균연령:", f_age_median)
```

```python
m_mask = titanic_sex_median["Sex"]=="male"
m_age_median = titanic_sex_median.loc[m_mask, "Age"].median()
print("남성 평균연령:", m_age_median)
```

부울 인덱싱으로 남녀별로 결측치를 각각의 중위값으로 대체한다.

```python
titanic_sex_median.loc[f_mask, 'Age'] = titanic_sex_median.loc[f_mask, 'Age'].fillna(f_age_median)
titanic_sex_median.loc[m_mask, 'Age'] = titanic_sex_median.loc[m_mask, 'Age'].fillna(m_age_median)
```

모든 결측치가 사라졌음을 확인한다.

```python
titanic_sex_median.info()
```

- 방식 2: `groupby()` 활용

먼저 타이타닉 데이터셋을 복제한다.

```python
titanic_sex_median = titanic.copy()
```

여성과 남성의 중위연령을 확인한다.

```python
titanic_sex_median.groupby('Sex')['Age'].median()
```

아래 코드는 성별에 따라 결측치를 각각의 중위값으로 대체한다.

```python
titanic_sex_median_age = titanic_sex_median.groupby('Sex')['Age'].apply(lambda y:y.fillna(y.median()))
titanic_sex_median_age
```

다중인덱스의 레벨 1에 위치한 `PassensgerId`를 기준으로 오름차순으로 정렬한다.

```python
titanic_sex_median_age = titanic_sex_median_age.sort_index(level=1)
titanic_sex_median_age
```

`Age` 특성의 값을 새롭게 지정한다.

```python
titanic_sex_median.loc[:, 'Age'] = titanic_sex_median_age.values
```

모든 결측치가 사라졌음을 확인한다.

```python
titanic_sex_median.info()
```

## 연습문제

`Age`의 결측치가 성별 중위연령으로 대체된 타이타닉 데이터셋을 이용한다.

```python
titanic = titanic_sex_median
titanic
```

**문제 1**

(1) 891명의 10%를 무작위로 추출했을 때의 성비율을 확인하는 코드르 작성하라.

```python
# 코드를 작성한다.
# 필요하면 코드셀 또는 텍스트셀을 추가할 수 있습니다.
random_sample = titanic.sample(frac=0.1, random_state=42)
random_sample = random_sample.groupby('Sex', observed = False).count()
random_sample = random_sample / random_sample.sum()
random_sample
```

(2) 891명의 10%를 성비율을 반영하면서 층화표집으로 추출했을 때의 성비율을 확인하는 코드를 작성하라.

```python
# 코드를 작성한다.
# 필요하면 코드셀 또는 텍스트셀을 추가할 수 있습니다.
stratified_sample = titanic.groupby('Sex', observed=True, group_keys=True).apply(lambda x: x.sample(frac=0.1, random_state=42), include_groups = False)
stratified_sample = stratified_sample.reset_index(level=0)
stratified_sample = stratified_sample.groupby('Sex', observed = False).count()
stratified_sample = stratified_sample / stratified_sample.sum()
stratified_sample
```

(3) 무작위 추출과 층화표집의 결과를 비교하는 데이터프레임을 생성하는 코드를 작성하라.

```python
# 코드를 작성한다.
# 필요하면 코드셀 또는 텍스트셀을 추가할 수 있습니다.
total_sample = titanic.groupby('Sex').count() / len(titanic)
df = pd.concat([total_sample.iloc[:, [1]],
               random_sample.iloc[:, [1]],
                stratified_sample.iloc[:,[1]]],axis=1)
df.columns = ['전체','무작위','층화']
df
```

```python
df['층화 오차율'] = (df['층화'] / df['전체'] - 1)
df['무작위 오차율'] = (df['무작위'] / df['전체'] - 1)
df
```

**문제 2**

(1) `Age` 특성을 10살 단위로 구분하는 연령구간을 지정하여 `Age_Bucket` 특성으로 추가하는 코드를 작성하라.

```python
# 코드를 작성한다.
# 필요하면 코드셀 또는 텍스트셀을 추가할 수 있습니다.
titanic['Age_Bucket'] = (titanic['Age'] // 10 * 10).astype('int64')
titanic
```

(2) 추가된 연령구간 정보를 활용하여 891명의 10%를 층화표집하는 코드를 작성하라.

```python
# 코드를 작성한다.
# 필요하면 코드셀 또는 텍스트셀을 추가할 수 있습니다.
stratified_sample = titanic.groupby('Age_Bucket' , observed = True , group_keys=True).apply(lambda x: x.sample(frac=0.1, random_state=42), include_groups=False)
stratified_sample = stratified_sample.reset_index(level=0)
stratified_sample
```

(3) 층화표집의 결과와 무작위 추출의 결과를 비교하는 데이터프레임을 생성하는 코드를 작성하라.

```python
# 코드를 작성한다.
# 필요하면 코드셀 또는 텍스트셀을 추가할 수 있습니다.
random_sample = titanic.sample(frac = 0.1 , random_state=42)
random_sample = random_sample.groupby('Age_Bucket').count() / len(random_sample)
random_sample
```

```python
stratified_sample_age = stratified_sample.groupby('Age_Bucket', observed=False).count() / len(stratified_sample)
stratified_sample_age
```

```python
total_sample = titanic.groupby('Age_Bucket').count() / len(titanic)
```

```python
df = pd.concat([total_sample.iloc[:,[1]],
                random_sample.iloc[:,[1]],
                stratified_sample_age.iloc[:,[1]]],axis=1)
df.columns = ['전체','무작위','층화']
df
```

```python
df['층화 오차율'] = (df['층화'] / df['전체'] - 1)
df['무작위 오차율'] = (df['무작위'] / df['전체'] - 1)
df
```
